```python
from keras.datasets import imdb
#from tanserkeras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Embedding(10000, 2,50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
from keras.layers import Input

model = Sequential([
    Input(shape=(50,)),  # Explicit input shape
    Embedding(input_dim=40000, output_dim=2),
    SimpleRNN(32),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# Model compilation (fixed)
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']  # Correct metrics format
)

# Model training (fixed)
history = model.fit(
    X_train, 
    y_train,
    epochs=5,  # Use actual number
    validation_data=(X_test, y_test)  # Correct syntax
)
```
