### Worling with dates

```python
# convet datetime dataype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### extract year

```python
date['date_year']= date['date'].dt.year
date.head()
```

### extract months


```python
date['date_month']= date['date'].dt.month
date.head()
```

```python
date['date_monthname']= date['date'].dt.month_name()
date.head()
```

```python
date['day']= date['date'].dt.day
date.head()
```

```python
date['day_name']= date['date'].dt.day_name()
date.head()
```

### is it weekend

```python
date['day_is_Weekend']= np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```
