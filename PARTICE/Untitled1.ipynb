```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('bank-additional-full.csv',sep=';')
df.sample(10)
```

```python
x = df.iloc[:,[1,3]]
```

```python
y = df.iloc[:,-1]
```

```python
xtrain , xtest, ytrain , ytest = train_test_split(x,y,test_size=0.75, random_state=42)
```

```python
od  = OrdinalEncoder(categories=[["management","admin.","technician","entrepreneur","self-employed","blue-collar","services","housemaid","retired","unemployed","student","unknown"],
['university.degree','professional.course','high.school','basic.9y','basic.6y','basic.4y','unknown','illiterate']])
```

```python
od.fit(xtrain)
```

```python
xtrain = od.transform(xtrain)
xtest = od.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression  
# 1. Create th
model = LogisticRegression()

# 2. Train the model
model.fit(xtrain, ytrain)

# 3. Predict on test data
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```
